```python
import glob, re, pandas as pd, fitz, xlsxwriter, string, os, sys
from pathlib import Path

# Struktur kata kunci dan grupnya
KEYWORDS_GROUPS = {
    "Global Reporting Initiative": "GRI Disclosure",
    "GRI Standard": "GRI Disclosure",
    "GRI Index": "GRI Disclosure",
    "SASB Standard": "SASB Disclosure",
    "SASB Disclosure": "SASB Disclosure",
    "SASB Index": "SASB Disclosure",
    "TCFD": "TCFD Disclosure",
    "TCFD Disclosure": "TCFD Disclosure",
    "Task Force on Climate-related Financial Disclosures": "TCFD Disclosure",
    "Value Creation": "Framework",
    "Sustainability Strategy": "Framework",
    "Our Progress": "Roadmap",
    "Our Sustainability Goals & Progress": "Roadmap",
    "Net-Zero Roadmap": "Roadmap",
    "materiality matrix": "Materiality Matrix"
}

# Fungsi untuk membersihkan teks dari karakter ilegal
def remove_illegal_characters(text):
    printable = set(string.printable)
    text = "".join(filter(lambda x: x in printable, text))
    illegal_chars = r"[\000-\010]|[\013-\014]|[\016-\037]"
    return re.sub(illegal_chars, "", text)

# Fungsi untuk mendapatkan semua file PDF dalam suatu folder
def get_all_pdf_files(folder_path):
    return [os.path.join(root, file) for root, _, files in os.walk(folder_path) for file in files if file.endswith(".pdf")]

# Fungsi untuk mencari kata kunci dalam file PDF
def search_keywords_in_pdf(file_path, df_results, tahun=None):
    file_name = os.path.basename(file_path)
    ticker = file_name.split("_")[0]
    found = False
    try:
        doc = fitz.open(file_path)
        for page_num, page in enumerate(doc):
            text = page.get_text()
            text = remove_illegal_characters(text)
            for keyword in KEYWORDS_GROUPS.keys():
                if re.search(r"\b" + re.escape(keyword) + r"\b[\s.,;:!?)]*", text, re.IGNORECASE):
                    found = True
                    new_row = pd.DataFrame({
                        "Ticker": [ticker],
                        "Tahun": [tahun],  # Menggunakan nilai default jika tahun tidak disediakan
                        "File Path": [file_path],
                        "Keyword": [keyword],
                        "Page": [page_num + 1],
                        "Text": [text]
                    })
                    df_results = pd.concat([df_results, new_row], ignore_index=True)
        if not found:  # Jika tidak ada kata kunci yang ditemukan dalam seluruh dokumen
            new_row = pd.DataFrame({
                "Ticker": [ticker],
                "Tahun": [tahun],  # Menggunakan nilai default jika tahun tidak disediakan
                "File Path": [file_path],
                "Keyword": [None],
                "Page": [None],
                "Text": [None]
            })
            df_results = pd.concat([df_results, new_row], ignore_index=True)
        doc.close()
    except Exception as e:
        print(f"Error processing PDF: {file_path}, Error: {str(e)}")
    return df_results

# Spesifik folder yang ingin di-scrap
folder_path = Path("./TOP COMPANIES BATCH 1")
output_folder_path = Path("./TOP COMPANIES BATCH 1/Scrap Done")

output_folder_path.mkdir(exist_ok=True)

# Initialize DataFrame for results
df_results = pd.DataFrame(columns=["Ticker", "Tahun", "File Path", "Keyword", "Page", "Text"])

# Get all PDF files in the specified folder
pdf_files = get_all_pdf_files(folder_path)
for file_path in pdf_files:
    df_results = search_keywords_in_pdf(file_path, df_results)  # Argumen tahun dihapus
    print(f"Processed file: {file_path}")

# Save the results to an Excel file
excel_path = output_folder_path / f"Green Bond_TOP COMPANIES BATCH 1.xlsx"
df_results.to_excel(excel_path, index=False)
print("Results saved to Excel.")
```
